# 构造人造的数据集

````python
def synthetic_data(w, b, num_example):
    """生成 y = Xw + b + 噪音"""
    # torch.normal(2, 3, size=(1, 4)) api中的规定X是由num_example个w维的向量构成的，现在看来是行向量
    # X 是均值为0标准差为1
    X = torch.normal(0, 1, (num_example, len(w))) 
    # 这里已经是矩阵与向量的乘积了
    y = torch.matmul(X, w) + b
    # 在此基础上加上一些噪音，y.shape与y的形状一样这样的话每一个y都会有高斯分布的噪音
    # reshape中-1表示自动计算，1表示固定，即列向量为1
    y += torch.normal(0, 0.01, y.shape)
    return X, y.reshape((-1, 1))

# 调用函数生成数据
true_w = torch.tensor([2, -3.4])
true_b = 4.2
features, labels = synthetic_data(true_w, true_b, 1000)
````

# 读取数据集

````python
# 每一次读取一个批量
def data_iter(batch_size, features, labels):
    # 这里的features的个数就是样本个数
    num_examples = len(features)
    # 得到下标，indices是index的复数
    indices = list(range(num_examples))
    # 洗牌，也就是打乱索引
    random.shuffle(indices)
    # 开始 结束 步长
    for i in range(0, num_examples, batch_size):
        # 得到一个随机的tensor，包含10个下标
        batch_indices = torch.tensor(
            # 取刚才随机的list中i到i+10的下标
            indices[i : min(i + batch_size, num_examples)]
        )
        # 返回一个迭代器，这个生成器函数每次都会停留在yield处等待下次执行的时候再生成一次
        yield features[batch_indices], labels[batch_indices]
````

在Python中，`yield` 是用于定义生成器函数（generator function）的关键字。生成器函数是一种特殊类型的函数，可以返回一个迭代器（iterator），并且能够在迭代过程中保存状态。

当函数中包含了 `yield` 语句时，它会被认为是一个生成器函数。生成器函数与普通函数的主要区别在于，当调用生成器函数时，并不会立即执行函数体内的所有代码，而是返回一个生成器对象。每次迭代生成器对象时，函数会从上次 `yield` 语句停止的地方继续执行，直到再次遇到下一个 `yield` 语句，然后再次停止。

以下是一个简单的示例，展示了生成器函数的用法：

```python
def my_generator():
    yield 1
    yield 2
    yield 3

# 调用生成器函数，返回一个生成器对象
gen = my_generator()

# 通过迭代生成器对象，获取每个 yield 语句的值
print(next(gen))  # 输出: 1
print(next(gen))  # 输出: 2
print(next(gen))  # 输出: 3
```

在上述示例中，`my_generator()` 是一个生成器函数。当调用它时，它返回一个生成器对象 `gen`。每次调用 `next(gen)` 时，生成器会从上次停止的地方继续执行，然后返回下一个 `yield` 语句的值。

生成器函数的优点是可以按需生成值，而不需要一次性生成所有的值，这在处理大量数据或无限序列时非常有用。此外，生成器函数还可以通过循环进行迭代，而不需要事先知道生成的值的数量。

# 初始化模型参数

````python
w = torch.normal(0, 0.01, size = (2,1), requires_grad = True)
b = torch.zeros(1, requires_grad = True)
````

````python
# 在这里实际上就是一个标量
torch.zeros(1)
````

# 定义模型

````python
# 下面的save的意思是将写的函数保存在d2l的包中
def linreg(X, w, b): #@save
    """线性回归模型"""
    return torch.matmul(X, w) + b
````

# 定义损失函数

````python
def squared_loss(y_hat, y): #@save
    """均方误差"""
    # 均方误差最后是要求的是平方差的均值这里没有除以batch_size 而在下面的优化算法中进行了计算，效果一样
    # 下面做的**2都是按照元素去乘的
    return (y_hat - y.reshape(y_hat.shape)) ** 2 / 2
````

````python
# Test
y1 = torch.Tensor([1,2,3,4,5])
y2 = torch.Tensor([5,4,3,2,1])
(y1-y2)**2/2
````

# 定义优化算法

````python
def sgd(params, lr, batch_size): #@save
    """小批量随机梯度下降"""
    # 注意这个no_grad的写法这个就是个代码块
    with torch.no_grad():
        for param in params:
            param -= lr * param.grad / batch_size
            param.grad.zero_() # pytorch不会梯度清零所以必须要梯度清零
````

# 开始训练

````python
# 定义超参数
lr = 0.03
# 整个数据扫描3轮，但是每次只使用10个数据来更新
num_epochs = 3
# 下面的函数变量化这样可以很方便的切换模型
net = linreg
loss = squared_loss
````

````python
# 正式训练
for epoch in range(num_epochs):
    # 每一次取一个批量的数据来训练，这个for中返回的是X与y然后batch_size只是制定了每次返回的大小
    for X, y in data_iter(batch_size, features, labels):
        l = loss(net(X, w, b), y) # X和y的小批量损失
        # 计算得到的loss是一个(batch_size,1)，而不是一个标量。l中的所有元素被加到一起，所以要相加做梯度计算
        # 注意这里面就涉及到前面所说的隐式构造图的计算图技术，可以直接自动微分
        l.sum().backward()
        sgd([w, b], lr, batch_size) # 在上面已经写到了将计算的梯度存放在w与b中
    # 下面这段代码是大for中的
    with torch.no_grad(): # 如果一段代码无需计算梯度那么就放在这个代码段中，对，他就是个代码块
        train_l = loss(net(features, w, b), labels) # 实务上是要计算训练过一轮之后的loss
        print(f'epoch {epoch + 1}, loss{float(train_l.mean()):f}')
````

# 线性回归的简单实现 使用Pytorch

# 生成数据集

````python
import numpy as np
import torch
from torch.utils import data
from d2l import torch as d2l
````

````python
true_w = torch.tensor([2, -3.4])
true_b = 4.2
features, labels = d2l.synthetic_data(true_w, true_b, 1000)
````

# 读取数据集

````python
def load_array(data_arrays, batch_size, is_train = True): #@save
    """构造一个PyTorch数据迭代器"""
    dataset = data.TensorDataset(*data_arrays)
    # 从dataset中每一次随机 拿一个batch出来
    return data.DataLoader(dataset, batch_size, shuffle = is_train)
````

````python
batch_size = 10
# (features, labels) 合成一个list
data_iter = load_array((features, labels), batch_size)
next(iter(data_iter))
````

# 使用框架定义的一个层

````python
from torch import nn
# 相当于神经网络的线性层
net = nn.Sequential(nn.Linear(2, 1)) # 2,1分别代表输入维度与输出维度，之所以套在Sequential里面是为了兼容以后多层神经网络的情形
````

# 初始化模型参数

````python
net[0].weight.data.normal_(0, 0.01)
net[0].bias.data.fill_(0)
# 第0层，权重或者偏置，的数值初始化
````

# 定义损失函数

````python
loss = nn.MSELoss()
````

# 定义优化算法

````python
# 将参数传入SGD，与学习率传入
trainer = torch.optim.SGD(net.parameters(), lr = 0.03)
````

# 开始训练

````python
num_epochs = 3
for epoch in range(num_epochs):
    for X, y in data_iter:
        # net自带模型参数，无需显式的传入
        l = loss(net(X), y)
        trainer.zero_grad()
        l.backward() # 自动求sum
        trainer.step() # 进行模型的更新
    l = loss(net(features), labels)
    print(f'epoch {epoch + 1}, loss {l:f}')
````

````python
w = net[0].weight.data
print('w的估计误差：', true_w - w.reshape(true_w.shape))
b = net[0].bias.data
print('b的估计误差：', true_b - b)
````
