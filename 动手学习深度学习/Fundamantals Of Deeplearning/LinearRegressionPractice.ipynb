# 构造数据集

```python
# 根据y = wx + b，参数有W b主要是生成对应的x与y，当然y需要加上随机噪音
def synthetic_data(w,b,num):
    # a sequence of integers defining the shape of the output tensor.
    X = torch.normal(0,1,(num,len(w)))
    # y = wX + b + rand
    """
        计算矩阵与向量的乘积
        result = torch.matmul(matrix, vector)
    """
    y = torch.matmul(X,w) + b
    # 加上随机性，实际上也是加上一个随机性的向量，这里的方差小了一些
    y += torch.normal(0,0.01,y.shape)
    # (-1,1)自动计算，一列
    return X,y.reshape(-1,1)

# 生成数据，这里需要注意的是w也是一个tensor
w = torch.Tensor([-10,5])
b = 3
num = 1000
X,y = synthetic_data(w,b,num)
```

# 读取数据集

```python
# 主要是读取一个beach使用到了yield功能，由于返回的beach需要被再次遍历所以返回的是一个迭代器
def batch_iter(batch_size,X,y):
    # 生成一个随机的列表
    num = len(X)
    index = list(range(num))
    # shuffle
    random.shuffle(index)
    # 搞10个数据过来，起始结束步长
    for i in range(0, num, batch_size):
        # 批量的下标也是个tensor
        batch_index = torch.tensor(index[i:min(i+batch_size,num)])
        yield X[batch_index],y[batch_index]
```

# 初始化参数

```python
# 注意生成的参数与模型的初始参数是不一样的，同时需要注意的是这些参数是需要求导的  所以需要保留梯度
w = torch.normal(0,0.01,size = (2,1),requires_grad = True)
# b的初值设置成0
b = torch.zeros(1,requires_grad = True)
```

# 定义模型

```python
# 这就是个线性模型
def linReg(X,w,b):
    return torch.matmul(X,w) + b
```

# 定义损失函数

```python
# 平方损失
def squared_loss_d(y, y_hat):
    # 我尝试一下直接使用内积
    y_tmp = y-y_hat.reshape(y.shape)
    return torch.dot(y_tmp) / 2
```

```python
# 哈德曼积然后求和在数值上等价于内积
# 平方损失
def squared_loss(y, y_hat):
    # 我尝试一下直接使用内积
    y_tmp = y - y_hat.reshape(y.shape)
    # 这里的**是乘方运算符
    return y_tmp ** 2 / 2
# return (y_hat - y.reshape(y_hat.shape)) ** 2 / 2
```

# 定义优化函数

```python
# 优化算法使用的是sgd
def sgd(params, lr, batch_size):
    # 实际上的主要任务就是跟新梯度，梯度蕴含在参数中
    # 梯度在训练的时候计算出来然后这里只是负责更新所以这里不需要使用bp计算梯度
    with torch.no_grad():
        for param in params:
            # 一定要注意除以batch
            param -= param.grad * lr / batch_size
            param.grad.zero_()
```

# 开始训练

```python
# 定义超参数
lr = 0.01
num_epochs = 5
batch_size = 10
# 相当于接口
net = linReg
loss = squared_loss
```

```python
# 正式训练，训练3次，每次拿上一个批去更新梯度
for epoch in range(num_epochs):
    for batch_data in batch_iter(batch_size,X,y):
        l = loss(y,net(X,w,b))
        # 计算梯度
        l.sum().backward()
        # 更新梯度
        sgd([w,b], lr, batch_size)
    # 上面的backward已经计算过梯度了下面就不需要计算梯度了
    with torch.no_grad():
        # 实际上是要打印一下梯度
        train_l = loss(y,net(X,w,b))
        print(f'epoch {epoch + 1}, loss{float(train_l.mean()):f}')
```
